```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS worldbank.data_2015 AS
SELECT
  g.country_name AS country_name,
	g.`2015` AS gdp,
  e.`2015` AS gov_edu_exp,
	l.`2015` AS literacy_rate,
	p.`2015` AS primary_enrollment,
	t.`2015` AS pupil_teacher_ratio,
	s.`2015` AS secondary_enrollment
FROM worldbank.gdp g
JOIN worldbank.gov_edu_exp e
  ON g.country_name = e.country_name
JOIN worldbank.literacy_rate l
  ON g.country_name = l.country_name
JOIN worldbank.primary_enrollment p
  ON g.country_name = p.country_name
JOIN worldbank.pupil_teacher_ratio t
  ON g.country_name = t.country_name
JOIN worldbank.secondary_enrollment s
  ON g.country_name = s.country_name
""")
```

```python
query = """
SELECT 
    country_name,
    gdp,
    gov_edu_exp,
    primary_enrollment,
    secondary_enrollment,
	pupil_teacher_ratio
FROM worldbank.data_2015
WHERE 
	country_name IS NOT NULL AND
	gdp IS NOT NULL AND
	gov_edu_exp IS NOT NULL AND
	primary_enrollment IS NOT NULL AND
	secondary_enrollment IS NOT NULL AND
	pupil_teacher_ratio IS NOT NULL
"""

df = pd.read_sql(query, conn)
df.head()
```
